```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Data Set 1 - Pie chart 
titles = ["Engineer", "Senior Staff", "Staff", "Senior Engineer", "Assitant Engineer", "Technique Leader", "Manager"]
titles_count = [19967, 19583, 6086, 5621, 2892, 2704, 6]
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue","orange","blue", "yellow" ]
explode = (0, 0, 0, 0.05, 0, 0,0)

plt.title("Potential Mentorees (4 year birth date Span 1962-1965)")
plt.pie(titles_count, explode=explode, colors=colors,
        labels=titles, autopct="%1.2f%%", startangle=0)
plt.axis("equal")
plt.savefig('Potential Mentorees_4 year birth.png')
plt.show()
```
